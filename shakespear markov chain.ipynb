```python
from nltk.corpus import stopwords
set1=set(stopwords.words('english'))
```

```python
from nltk.tokenize import word_tokenize
def stop_words1(col):
    text=col[0]
    word_tokens = word_tokenize(text)
    filtered_sentence = [w for w in word_tokens if not w in set1]
    return filtered_sentence
df1['PlayerLine']=df[['PlayerLine']].apply(stop_words1,axis=1)
```

```python
def con_tolower(col):
    text=col[0]
    for i in range(0,len(text)):
        text[i] = text[i].lower()
    return text
df1['PlayerLine']=df1[['PlayerLine']].apply(con_tolower,axis=1)
```

```python
def rem_special1(col):
    text=col[0]
    text1=[]
    for i in range(0,len(text)-2):
        if text[i] != '':
            text1.append(text[i])
    return text1
df1['PlayerLine']=df1[['PlayerLine']].apply(rem_special1,axis=1)
```

```python
dic = {}
def fn(col):
    corpus = col[0]
    #print(corpus)
    for i in range(len(corpus)-1):
        #print(corpus[i]+' '+ corpus[i+1])
        if corpus[i] in dic.keys():
            dic[corpus[i]].append(corpus[i+1])
        else:
            dic[corpus[i]] = [corpus[i+1]]
```

```python
df1[['PlayerLine']].apply(fn,axis=1)
```

```python
from collections import Counter
```

```python
ndic={}
for i in dic.keys():
    row=dic[i]
    c=Counter(row)
    ndic[i]=c.most_common(1)[0][0]
```
